```python
import re
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from scipy.stats import pearsonr
from datetime import datetime
import seaborn as sns

data_path = './../../../data'
```

```python
rb_dir = os.path.join(data_path, 'RateBeer')
rev_with_scores_rb = pd.read_pickle(os.path.join(rb_dir, 'rev_w_scores.pkl'))
rev_with_scores_rb['date'] = pd.to_datetime(rev_with_scores_rb['date'])
beers_rb = pd.read_csv(os.path.join(rb_dir, 'beers_RB_clean.csv'))
users_rb = pd.read_csv(os.path.join(rb_dir, 'users_RB_clean.csv'))
rev_with_scores_rb.columns = ['aroma_rt' if i == 10 else col for i, col in enumerate(rev_with_scores_rb.columns)]
rev_with_scores_rb.columns = ['appearance_rt' if i == 9 else col for i, col in enumerate(rev_with_scores_rb.columns)]
```

```python
complete_beer = pd.merge(beers_rb, rev_with_scores_rb, on='beer_id')
```

```python
col_to_keep = ['style_x','flavor', 'aroma', 'mouthfeel', 'brewing', 'technical', 'appearance', 'judgment','off_flavors', 'miscellaneous', 'expertness_score']

beer_gr = complete_beer[col_to_keep].groupby('style_x')
```

## Analysis on IPA

```python
rev_Ipa = complete_beer[complete_beer['style_x']=='India Pale Ale (IPA)']
col_to_keep = ['user_id','stems','flavor', 'aroma', 'mouthfeel', 'brewing', 'technical', 'appearance', 'judgment','off_flavors', 'miscellaneous', 'expertness_score']

rev_Ipa = rev_Ipa[col_to_keep]
complete_IPA = pd.merge(rev_Ipa, users_rb[['location', 'user_id']], on='user_id')
```

```python
loc_size = complete_IPA.groupby('location').size().sort_values()
to_plot = loc_size[loc_size > 5000]
fig  = plt.figure(figsize=(17,7))
plt.bar(to_plot.index, to_plot)

plt.xticks(rotation=90)
plt.show()
```

```python
loc_size[loc_size > 50]
```

```python
sel_countries = ['Canada', 'Denmark', 'England', 'United States, California', 'Germany', 'Poland']
```

```python
loc_score = complete_IPA[complete_IPA['location'].isin(sel_countries)][['location', 'expertness_score']].groupby('location').mean()
loc_score = loc_score.sort_values(by='expertness_score')
plt.title('IPA Expertness_score')
plt.bar(loc_score.index, loc_score.expertness_score)
plt.xticks(rotation=45)
plt.show()
```

```python
loc_score
```

```python
def compute_occ(df):
    occ = {}
    for x in df:
        for term in x:
            if term in occ.keys():
                occ[term]+= 1
            else: 
                occ[term]= 1
    return pd.Series(occ)
```

```python
#loc_size = complete_IPA[complete_IPA['location'] in sel_countries].groupby('location')
count_word = pd.DataFrame()
for c in sel_countries:
    count_word[c] = compute_occ(complete_IPA[complete_IPA['location']==c].stems)
    #print(complete_IPA[complete_IPA['location']==c].stems.value_counts())
    ## TODO fare un dizionario che conta le occorenze e poi plottarlo

count_word =count_word.fillna(0)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 6)) 


fig.suptitle("IPA Most frequent words per country", fontsize=16)

for i, c in enumerate(count_word.columns):
    sel_c = count_word[c].sort_values(ascending =False).head(20)
    axes[i//2, i%2].bar(sel_c.index, sel_c, color='skyblue')
    plt.setp(axes[i//2, i%2].get_xticklabels(), rotation=45, ha='right')
    axes[i//2, i%2].set_title(c)
    
plt.tight_layout()  
plt.show()

```
