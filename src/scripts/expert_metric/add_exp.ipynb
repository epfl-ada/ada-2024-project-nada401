```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

import sys
sys.path.append('../')
from helpers import *
```

```python
data_path = './../../../data'
advocate_dir = os.path.join(data_path, 'BeerAdvocate')
stems_ba = pd.read_pickle(os.path.join(advocate_dir, 'reviews_with_exp_stems.pkl'))
stems_ba.columns = ['appearance_rt' if i == 9 else col for i, col in enumerate(stems_ba.columns)]
expert_categories = get_exp_categories(data_path)
expert_terms_stemmed = get_exp_stems(data_path)
```

```python
def score_df(df, expert_categories, expert_terms_stemmed):
    for criterion in expert_categories:
        df[criterion] = df['stems'].apply(lambda cell: len(list(set(cell) & set(expert_terms_stemmed[criterion]))))
    df['expertness_score'] = df[expert_categories].sum(axis=1)
    return df
```

```python
import time

start_time = time.time()

df_ba = score_df(stems_ba, expert_categories, expert_terms_stemmed)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df_ba.to_pickle(os.path.join(advocate_dir, 'rev_w_scores.pkl'))
df_ba.head()
```
