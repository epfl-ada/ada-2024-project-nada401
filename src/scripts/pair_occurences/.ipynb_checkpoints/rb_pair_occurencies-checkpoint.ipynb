# Pair occurence analysis

```python
import re
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from scipy.stats import pearsonr
from datetime import datetime
import seaborn as sns
import time

data_path = './../../../data'
```

```python
rb_dir = os.path.join(data_path, 'RateBeer')
rev_with_scores_rb = pd.read_pickle(os.path.join(rb_dir, 'rev_w_scores.pkl'))
rev_with_scores_rb['date'] = pd.to_datetime(rev_with_scores_rb['date'])
beers_rb = pd.read_csv(os.path.join(rb_dir, 'beers_RB_clean.csv'))
users_rb = pd.read_csv(os.path.join(rb_dir, 'users_RB_clean.csv'))
rev_with_scores_rb.columns = ['aroma_rt' if i == 10 else col for i, col in enumerate(rev_with_scores_rb.columns)]
rev_with_scores_rb.columns = ['appearance_rt' if i == 9 else col for i, col in enumerate(rev_with_scores_rb.columns)]
```

```python
def compute_pair_occ(df):
    pair_occ = {}
    for x in df:
        if len(x)<2:
            continue
        for i in range(len(x)-1):
            j = i+1
            while j < len(x):
                pair = x[i]+', '+x[j] if x[i] < x[j] else x[j]+', '+x[i]
                if pair in pair_occ.keys():
                    pair_occ[pair]+= 1
                else: 
                    pair_occ[pair]= 1
                j+=1
    return pair_occ
```

```python
start_time = time.time()

pair_occ = compute_pair_occ(rev_with_scores_rb.stems)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
pair_occ = sorted(pair_occ.items(), key=lambda x:x[1] , reverse=True)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
expert_terms_stemmed = {}
for category, terms in expert_terms.items():
    tokens = [word_tokenize(term.lower()) for term in terms]
    
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word[0]) for word in tokens]
    expert_terms_stemmed[category] =  stemmed_tokens

```

```python
def compute_pair_occ_cat(df, expert_terms_stemmed):
    pair_occ_cat = {cat: {} for cat in expert_terms_stemmed.keys()}
    for x in df:
        if len(x)<2:
            continue
        for i in range(len(x)-1):
            for cat in expert_terms_stemmed.keys():
                if x[i] in expert_terms_stemmed[cat]:
                    j = i+1
                    while j < len(x):
                        if x[j] in expert_terms_stemmed[cat]:
                            pair = x[i]+', '+x[j] if x[i] < x[j] else x[j]+', '+x[i]
                            if pair in pair_occ_cat[cat].keys():
                                pair_occ_cat[cat][pair]+= 1
                            else: 
                                pair_occ_cat[cat][pair]= 1
                        j+=1
    return pair_occ_cat
```

```python
star_time = time.time()
pair_occ_cat = compute_pair_occ_cat(rev_with_scores_rb.stems, expert_terms_stemmed)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def print_50(dictionary, title='', first = True, semilog=False):
    if first:
        x_values = [t[0] for t in dictionary[:50]]
        y_values = [t[1] for t in dictionary[:50]]
    else:
        x_values = [t[0] for t in dictionary[-50:]]
        y_values = [t[1] for t in dictionary[-50:]]
    
    # Create the plot
    plt.figure(figsize=(12, 8))
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
    if semilog:
        plt.yscale('log')
    plt.title(title + '50' +(' most ' if first else ' least ')+ 'common pairs')
    plt.xticks(rotation=90)
    plt.show()
```

```python
flavor_pair_occ = sorted(pair_occ_cat['flavor'].items(), key=lambda x:x[1] , reverse=True)
print("Total pairs: %d" %(len(flavor_pair_occ)))
print_50(flavor_pair_occ, title='Flavor category: ', first=True, semilog=True)
print_50(flavor_pair_occ, title='Flavor category: ', first=False)
```

```python
aroma_pair_occ = sorted(pair_occ_cat['aroma'].items(), key=lambda x:x[1] , reverse=True)
print("Total pairs: %d" %(len(aroma_pair_occ)))
print_50(aroma_pair_occ, title='Aroma category: ', first=True, semilog=True)
print_50(aroma_pair_occ, title='Aroma category: ', first=False)
```

```python
mouthfeel_pair_occ = sorted(pair_occ_cat['mouthfeel'].items(), key=lambda x:x[1] , reverse=True)
print("Total pairs: %d" %(len(mouthfeel_pair_occ)))
title = 'Mouthfeel category: '
print_50(mouthfeel_pair_occ, title=title, first=True, semilog=True)
print_50(mouthfeel_pair_occ, title=title, first=False)
```
